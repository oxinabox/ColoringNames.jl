```julia
using ColoringNames
using Plots
```

```julia
cldata = load_munroe_data(dev_as_train=false, dev_as_test=true)
```

```julia
noml_unsmoothed = TermToColorDistributionEmpirical(256)
train!(noml_unsmoothed, cldata, Val{false}())
```

```julia
noml_smoothed = TermToColorDistributionEmpirical(256)
train!(noml_smoothed, cldata, Val{true}())
```

```julia
plot_query(noml_unsmoothed,  "brownish green")|> display
plot_query(noml_smoothed,  "brownish green")|> display
plot_query(noml_unsmoothed,  "brown")|> display
plot_query(noml_smoothed,  "brown")|> display
plot_query(noml_unsmoothed,  "brownish")|> display
plot_query(noml_smoothed,  "brownish")|> display
plot_query(noml_unsmoothed,  "greenish")|> display
plot_query(noml_smoothed,  "greenish")|> display
plot_query(noml_unsmoothed,  "greenish brown")|> display
plot_query(noml_smoothed,  "greenish brown")|> display
```
